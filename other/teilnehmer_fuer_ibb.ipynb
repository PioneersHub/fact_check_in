```python
import json
from datetime import datetime

speakers_days = json.load(open("/Users/hendorf/Documents/SyncThing/Vorstand/2023/Kongressfonds Berlin/Eingereichte Unterlagen/Teilnehmer/raw data/pyconde-pydata-berlin-2023_sessions.json"))
speaker_days = {}
for o in speakers_days:
    day = datetime.strptime(o["Start"], "%Y-%m-%dT%H:%M:%S+00:00")
    for s in o["Speaker names"]:
        try:
            speaker_days[day.weekday()].add(s)
        except KeyError:
            speaker_days[day.weekday()] = set([s])
speaker_days   
```

```python
all_infos_bcc["Organisation"] = all_infos_bcc["Organisation"].fillna("ZZZ-kA.").map(lambda x: x.strip())
all_infos_bcc["Land"] = all_infos_bcc["Land"].fillna("ZZZ-kA.").map(lambda x: x.strip())
all_infos_bcc["Stadt"] = all_infos_bcc["Stadt"].fillna("ZZZ-kA.").map(lambda x: x.strip())
```

```python
all_infos_bcc["sort"] = 0
all_infos_bcc.loc[(
                  (all_infos_bcc["Teilnehmende Person"] == "ZZZ-kA.") | 
                  (all_infos_bcc["Organisation"] == "ZZZ-kA.") | 
                  (all_infos_bcc["Land"] == "ZZZ-kA.") | 
                  (all_infos_bcc["Stadt"] == "ZZZ-kA.")
), "sort"] = 1
```

```python
def remove_random(df, number_to_select, wkday):
    sel = random.sample(list(df.index), number_to_select)
    sampled = df.loc[sel]
    # make sure speakers on days are in selection for day
    delta = set(speaker_days[wkday]) - set(sampled["Ticket Full Name"])
    if delta:
        add_again = df[df["Ticket Full Name"].isin(delta)]
        if add_again.size:
            sampled = df.loc[sel + list(add_again.index)]
    return sampled
    
```

```python
df_day = all_infos_bcc[all_infos_bcc["Monday"]]
cnt = 1305
monday = remove_random(df_day, cnt, 0)
monday.describe()
```

```python
df_day = all_infos_bcc[all_infos_bcc["Tuesday"]]
cnt = 1200
tuesday = remove_random(df_day, cnt, 1)
tuesday.describe()
```

```python
df_day = all_infos_bcc[all_infos_bcc["Wednesday"]]
cnt = 1190
wednesday = remove_random(df_day, cnt, 2)
wednesday.describe()
```

```python
monday.sort_values(["sort"], inplace=True)
monday.index = range(1,monday.shape[0]+1)
tuesday.sort_values(["sort"], inplace=True)
tuesday.index = range(1,tuesday.shape[0]+1)
wednesday.sort_values(["sort"], inplace=True)
wednesday.index = range(1,wednesday.shape[0]+1)


monday[["Teilnehmende Person", "Organisation", "Land", "Stadt", "Art der Teilnahme"]].to_excel("Teilnehmerliste-monday.xlsx", index=True)
tuesday[["Teilnehmende Person", "Organisation", "Land", "Stadt", "Art der Teilnahme"]].to_excel("Teilnehmerliste-tuesday.xlsx", index=True)
wednesday[["Teilnehmende Person", "Organisation", "Land", "Stadt", "Art der Teilnahme"]].to_excel("Teilnehmerliste-wednesday.xlsx", index=True)
```
